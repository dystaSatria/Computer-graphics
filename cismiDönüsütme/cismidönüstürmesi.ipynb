```python
import matplotlib.pyplot as plt
import numpy as np

# Verilen noktaların koordinatları
A = np.array([6, 2])
B = np.array([9, 2])
C = np.array([6, 4])
P = np.array([10, 10])

# Döndürme işlemi için dönüş matrisi oluşturma fonksiyonu
def rotate_point(point, angle, center):
    angle_rad = np.radians(angle)
    rotation_matrix = np.array([
        [np.cos(angle_rad), -np.sin(angle_rad)],
        [np.sin(angle_rad), np.cos(angle_rad)]
    ])
    return np.dot(rotation_matrix, point - center) + center

# Saat yönünde 70 derece döndürme işlemi
angle = 70
center = P
A_rotated = rotate_point(A, angle, center)
B_rotated = rotate_point(B, angle, center)
C_rotated = rotate_point(C, angle, center)
P_rotated = rotate_point(P, angle, center)

# Y ekseninde -12 birim öteleme işlemi
offset = -12
A_offset = A_rotated + [0, offset]
B_offset = B_rotated + [0, offset]
C_offset = C_rotated + [0, offset]
P_offset = P_rotated + [0, offset]

# Çizim
fig, ax = plt.subplots()

# Koordinat eksenlerini çizme
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)

# Noktaları ve üçgeni çizme
ax.plot([A_offset[0], B_offset[0]], [A_offset[1], B_offset[1]], 'k-')
ax.plot([B_offset[0], C_offset[0]], [B_offset[1], C_offset[1]], 'k-')
ax.plot([C_offset[0], A_offset[0]], [C_offset[1], A_offset[1]], 'k-')
ax.plot(P_offset[0], P_offset[1], 'o', color='red')  # P noktasını çizme
ax.annotate('P', (P_offset[0], P_offset[1]), textcoords="offset points", xytext=(-10,-10), ha='center')

# Eksenleri ve çizim aralığını ayarlama
plt.xlim(-15, 20)
plt.ylim(-15, 20)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(visible=True)
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.title('Döndürülmüş ve Ötelenmiş Üçgen')

# Çizimi gösterme
plt.show()
```
